```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt
```

```python

data= pd.read_csv("C:/Users/GRIET/Downloads/ToyotaCorolla.csv")
data
```

```python
print("SCATTER PLOT")
data.dropna(axis=1,inplace=True)
plt.scatter(data["age"],data["price"],c="red")
plt.title= "scatter plot of age vs price"
plt.xlabel("age")
plt.ylabel("price")
plt.show()
```

```python
print("HISTOGRAM")
plt.hist(data['km'],color='red',edgecolor='white',bins=5)

plt.title="histogram of km"

plt.xlabel('KM')

plt.ylabel('frequency')

plt.show()
```

```python
print("BAR GRAGH")
count=[979,120,12]

fuelType=('Petrol','Diesel','CNG')

index=np.arange(len(fuelType))

plt.bar(index,count,color=('red','blue','green'))

plt.title="bar plot of toyota corolla"

plt.xlabel('FuelType')

plt.ylabel('Frequency')

plt.xticks(index,fuelType,rotation=45)

plt.show()
```
